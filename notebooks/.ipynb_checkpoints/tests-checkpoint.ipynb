```python
import easyocr

reader = easyocr.Reader(['en'], gpu=False)
```

```python
# VER DOCUMENTAO PARA MELHORAR ISSO ADICIONANDO UNS PARMETROS
results = reader.readtext(img_fns[11])
results[0]
```

```python
df_word_boxes = pd.DataFrame(results, columns=['bbox', 'text', 'conf'])
```

```python
reader = easyocr.Reader(['en'], gpu=False)

dfs = []

for img in tqdm(img_fns[:25]):
    results = reader.readtext(img)
    img_id = img.split('/')[-1].split('\\')[-1].split('.')[0]
    img_df = pd.DataFrame(results, columns=['bbox', 'text', 'conf'])
    img_df['img_id'] = img_id
    dfs.append(img_df)
    
easyocr_df = pd.concat(dfs)
easyocr_df.reset_index(inplace=True)
easyocr_df
```
